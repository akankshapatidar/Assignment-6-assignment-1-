```python
##  TASK 1. Create a JSON file (employee.json) containing employee information of minimum 5 employees.
#   Each employee information consists of Name, DOB, Height, City, State. Write a python program 
#   that reads this information from the JSON file and saves the information into a list of objects
#   of Employee class. Finally print the list of the Employee objects.


import json

employees = [
    {
        "name": "Aakanksha Patidar",
        "DOB": "23-09-1996",
        "height": "5'2",
        "city": "Indore",
        "state": "Madhay pradesh\n",
    },
    
    {
        "name": "Abhilash Dhanoriya",
        "DOB": "28-12-1995",
        "height": "5'11",
        "city": "Bhopal",
        "state": "Madhay Pradesh\n",
    },
    
    {
        "name": "Anand Maladhari",
        "DOB": "25-01-1989",
        "height": "6'4",
        "city": "Chennai",
        "state": "Tamil Nadu\n",
    },
    
    {
        "name": "Shubham Pawar",
        "DOB": "04-07-1997",
        "height": "5'5",
        "city": "Bangalore",
        "state": "Karnataka\n",
    },
    
    {
        "name": "Tanishk Sharma",
        "DOB": "21-06-1997",
        "height": "5'6",
        "city": "Hyderabad",
        "state": "Telangana State\n",
    },
    
    {
        "name": "Ashutosh Pandey",
        "DOB": "21-01-1998",
        "height": "5'10",
        "city": "Pune",
        "state": "Maharastra\n",
    },
    
    {
        "name": "Ashvini Dawar",
        "DOB": "26-06-1998",
        "height": "5'3",
        "city": "Indore",
        "state": "Madhya Pradesh\n",
    }

]


with open("employee.json", "w") as file:
    json.dump(employees, file)

class Employee:
    def __init__(self, name, DOB, height, city, state):
        self.name = name
        self.DOB = DOB
        self.height = height
        self.city = city
        self.state = state

def read_employee_data(filename):
    with open(filename, 'r') as f:
            data = json.load(f)
    employee_list = []
        
    for employee_data in data:
        name = employee_data['name']
        DOB = employee_data['DOB']
        height = employee_data['height']
        city = employee_data['city']
        state = employee_data['state']
        employee = Employee(name, DOB, height, city, state)
        employee_list.append(employee)
    return employee_list

def main():
    employee_list = read_employee_data('employee.json')
    for employee in employee_list:
        print(employee.name)
        print(employee.DOB)
        print(employee.height)
        print(employee.city)
        print(employee.state)

if __name__ == '__main__':
    main()
```
